```python
#funciones que arreglan el error de los registros con correccion de lecturas
def mean_con(x):
    i=-1
    if('AJUSTE LECTURA' in np.array(x.Categoría) ):
        i=np.where(np.array(x.Categoría)=='AJUSTE LECTURA')[0][-1]
    #return [((x.iloc[0,8]-x.iloc[i,8])/(x.iloc[0,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)-x.iloc[i,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)).days)*20,(x.iloc[0,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)-x.iloc[-1,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)).days]
    return ((x.iloc[0,8]-x.iloc[i,8])/(x.iloc[0,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)-x.iloc[i,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)).days)*20
def total_con(x):
    i=-1
    if('AJUSTE LECTURA' in np.array(x.Categoría) ):
        i=np.where(np.array(x.Categoría)=='AJUSTE LECTURA')[0][-1]
    return (x.iloc[0,8]-x.iloc[i,8])
```

```python
mean_consumptions=lectures.groupby(['Serie']).apply(lambda x:mean_con(x))
mean_consumptions=mean_consumptions.fillna(0)
```

```python
#dos series que aun tienes numero negativo en lecturas

series_error=mean_consumptions.loc[mean_consumptions<0].index
```

```python
total_consumptions=lectures.groupby(['Serie']).apply(lambda x:total_con(x))
```

```python
per=data['DIA DE CIERRE'].dt.to_period("M") 
results=pd.DataFrame(data.groupby(['SERIE',per]).apply(lambda x:sum(x.MONTO)),columns=['Maintainance cost'])
```

```python
results['mean_production']='0'
results['total_production']='0'
```

```python
results['mean_production']=list(mean_consumptions.loc[(results.index.get_level_values(0))])
results['total_production']=list(total_consumptions.loc[(results.index.get_level_values(0))])
results['model']=list(models.loc[(results.index.get_level_values(0))])
results['technician']=list(technicians.loc[(results.index.get_level_values(0))])


```

```python
results['cost_per_click']=results['Maintainance cost']/results['total_production']
```

```python
T_M_df=pd.DataFrame(total_consumptions.loc[indx]).join(models.loc[indx]).join(technicians.loc[indx])
#T_M_df=pd.DataFrame(total_consumptions).join(models).join(technicians)
T_M_df.columns=['total_consumption','Model','Technic']
df1=results[['model','technician','Maintainance cost']].loc[indx].groupby(['technician','model']).sum()
#df1=results[['model','technician','Maintainance cost']].groupby(['technician','model']).sum()
df2=T_M_df.groupby(['Technic','Model']).sum()
T_M_result=pd.concat([df1, df2], axis=1)
T_M_result['cost_per_click']=T_M_result['Maintainance cost']/T_M_result['total_consumption']
T_M_result
```
