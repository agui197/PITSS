```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel('analisis empleados.xlsx','data')
```

```python
data.head()
```

```python
lectures=pd.read_excel('analisis empleados.xlsx','lecturas_Data')
```

```python
lectures.head()
```

```python
errors=lectures.loc[list(map(lambda x:x=='AJUSTE LECTURA',lectures.Categoría))]
series_error=np.unique(errors.Serie)
```

```python
unique=pd.read_excel('analisis empleados.xlsx','unique',index_col=0)
models=unique.SERIE
technicians=unique.TECNICO
unique.head()
```

```python
mean_consumptions=lectures.groupby(['Serie']).apply(lambda x:[((x.iloc[0,8]-x.iloc[-1,8])/(x.iloc[0,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)-x.iloc[-1,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)).days)*20,(x.iloc[0,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)-x.iloc[-1,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)).days])
#mean_consuptions=lectures.groupby(['Serie']).apply(lambda x:((x.iloc[0,8]-x.iloc[-1,8])/(x.iloc[0,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)-x.iloc[-1,1].replace(hour = 0, minute = 0, second = 0, microsecond = 0)).days)*20)
mean_consumptions=mean_consumptions.fillna(0)
```

```python
total_consumptions=lectures.groupby(['Serie']).apply(lambda x:(x.iloc[0,8]-x.iloc[-1,8]))
```

```python
per=data['DIA DE CIERRE'].dt.to_period("M") 
results=pd.DataFrame(data.groupby(['SERIE',per]).apply(lambda x:sum(x.MONTO)),columns=['Maintainance cost'])
```

```python
results['mean_production']='0'
results['total_production']='0'
```

```python
results['mean_production']=list(mean_consumptions.loc[(results.index.get_level_values(0))])
results['total_production']=list(total_consumptions.loc[(results.index.get_level_values(0))])
results['serie']=list(models.loc[(results.index.get_level_values(0))])
results['technician']=list(technicians.loc[(results.index.get_level_values(0))])


```

```python
results['cost_per_click']=results['Maintainance cost']/results['total_production']
```
